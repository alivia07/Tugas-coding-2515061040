```python
# Membuat vektor
v = np.array([3, 4])

# Menghitung panjang (norma L2)
norm = np.linalg.norm(v)

# Menghitung unit vector
unit_v = v / norm

print("Vektor asli:", v)
print("Panjang vektor:", norm)
print("Unit vector:", unit_v)
print("Panjang unit vector:", np.linalg.norm(unit_v))
```
